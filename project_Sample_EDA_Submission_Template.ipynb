***REMOVING THE OUTLIER***

**+**-THE Price column is very important to us therefore we wanna remove the outilier so it give the proper result and helpfull for our future prediction

````python
#ploting the box plot for seeing the outlier
sns.boxplot(x=airbnb["price"],color="brown")
#showing the graph
plt.show()
````

````python
#calculating 25% and 75% of the price data
airbnb_25 = airbnb["price"].quantile(0.25)
airbnb_75 = airbnb["price"].quantile(0.75)
#itroducing iqr-(inter quartile range)
iqr = airbnb_75 - airbnb_25
#setting upper and lower limit
upper_limit = airbnb_75 + 1.5*iqr
lower_limit = airbnb_25 - 1.5*iqr
#creating the updated removed autlier dataframe
airbnb_new = airbnb[(airbnb["price"]<=upper_limit)] # not using lower limit as we have the min value 0 so in lower limit it goinng negetive which would be false
````

````python
#setting the graph style what to use in further analysis
#plt.style.available
plt.style.use('fivethirtyeight')
#plotting boxplot after removing the outlier
sns.boxplot(x=airbnb_new["price"],color="brown")
#showing graph
plt.show()
#NEw BOX PLOt
````
